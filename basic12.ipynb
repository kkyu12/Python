```python
f=0 # 변수 선언과 초기화
try: # 예외 에러 처리
    f=open("LICENSE.txt","rt") # 파일 열기
    while True: # 무한 반복문
        r=f.readline() # 한줄 가져오기
        if not r: # 라인이 없다면
            break # 반복문 탈출
        print(r) # 한줄씩 파일 내용 출력
except FileNotFoundError: # 예외 에러 처리
    print("파일없음") # 파일이 없을 경우 메세지 출력
finally: # 무조건 실행
    if(f): f.close() # 파일을 닫고
    print("프로그램 끝") # 메세지 출력
```

```python
# 100명의 국어,영어,수학 점수를 임의로 만든 후 csv형태의 파일의 기록하는 예제
import random as r # 랜덤 모듈 사용
f=0 # 변수 선언과 초기화
f=open("score.txt","wt") # 파일 쓰기로 파일 열기
# 아래와 같은 형태를 만들기 위하여 str.format 함수를 사용할 수 있다. 
#   one_person="{0:s},{1:d},{2:d},{3:d}\n"
#.format("홍길동"+str(i),r.randrange(0,100),r.randrange(0,100),r.randrange(0,100)) 
for i in range(100): # 100명
    one_person="홍길동"+str(i)+","+str(r.randrange(0,100))+","+str(r.randrange(0,100))+"," \
    +str(r.randrange(0,100))+"\n" # 이름+번호 , 국어점수 , 영어점수, 수학점수 만들기
    f.write(one_person) #한 줄씩 쓴다 
    print(one_person) # 한명씩 출력
f.close() # 파일 닫기
```

```python
f=0 # 변수 선언과 초기화
f=open("score.txt","rt")  # 파일 읽기로 파일 열기
while True: # 무한 반복문
    one_person=f.readline() # 한 줄씩 읽는다
    if not one_person: break # 읽을 내용이 없으면 루프탈출
    one_person=one_person.replace("\n","") # 줄 바꿈문자 없애기
    item=one_person.split(",") #콤마 구분자로 짜르기
    print("="*10) # 등호 10개 출력
    print("이름: ", item[0]) # 이름 출력
    print("국어: ", item[1]) # 국어 성적 출력
    print("영어: ", item[2]) # 영어 성적 출력
    print("수학: ", item[3]) # 수학 성적 출력
    print("="*10)  # 등호 10개 출력
f.close() # 파일 닫기
```

```python
import sqlite3 as s 
con=s.connect("my01.db") 
cursor = con.cursor()

cursor.execute("select * from examtable")

one_rec=cursor.fetchall()  #해당 데이터를 전부 가져옴

for field in one_rec:  #해당 데이터를 반복문을 통해 하나씩 처리 
    kor=field[2]
    eng=field[3] 
    mat=field[4]
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"\
          %(field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3))) 

cursor.close()
con.close()
```
